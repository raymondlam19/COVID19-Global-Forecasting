# EDA over worldwide

```python
worldwide = df_train.groupby(df_train.index).agg({'ConfirmedCases': 'sum', 'Fatalities': 'sum'})

worldwide.tail()
```

# Overview

```python
fig, ax = plt.subplots(3, 3, figsize=(18,12))

hk.plot(secondary_y='Fatalities', title='Hong Kong', ax=ax[0,0])
china.plot(secondary_y='Fatalities', title='China', ax=ax[0,1])
taiwan.plot(secondary_y='Fatalities', title='Taiwan', ax=ax[0,2])
korea.plot(secondary_y='Fatalities', title='Korea', ax=ax[1,0])
japan.plot(secondary_y='Fatalities', title='Japan', ax=ax[1,1])
malaysia.plot(secondary_y='Fatalities', title='Malaysia', ax=ax[1,2])
singapore.plot(secondary_y='Fatalities', title='Singapore', ax=ax[2,0])

plt.show()
```

**Observartion:**
* safe(stable): HK, China, Taiwan, Korea
* danger(rising): Japan, Malaysia, Singapore

```python
asia = {'Country': ['Hong Kong', 'China', 'Taiwan', 'Korea', 'Japan', 'Malaysia', 'Singapore'],
        'Population': [7496981, 1439323776, 23816775, 51269185, 126476461, 32318833, 5850342],
        'ConfirmedCases': [hk['ConfirmedCases'].iloc[-1], china['ConfirmedCases'].iloc[-1], taiwan['ConfirmedCases'].iloc[-1], korea['ConfirmedCases'].iloc[-1], japan['ConfirmedCases'].iloc[-1], malaysia['ConfirmedCases'].iloc[-1], singapore['ConfirmedCases'].iloc[-1]],
        'InfectedRate': [np.nan]*7,
        'Fatalities': [hk['Fatalities'].iloc[-1], china['Fatalities'].iloc[-1], taiwan['Fatalities'].iloc[-1], korea['Fatalities'].iloc[-1], japan['Fatalities'].iloc[-1], malaysia['Fatalities'].iloc[-1], singapore['Fatalities'].iloc[-1]],
        'DeathRate': [np.nan]*7,
       }

asia = pd.DataFrame(asia).set_index('Country')
asia['ConfirmedCases'] = asia['ConfirmedCases'].astype(int)
asia['Fatalities'] = asia['Fatalities'].astype(int)
asia['InfectedRate'] = asia['ConfirmedCases'] / asia['Population'] * 100
asia['DeathRate'] = asia['Fatalities'] / asia['ConfirmedCases'] * 100

asia
```

**Observartion:**
* highest infected rate: Singapore; lowest infected rate: Taiwan
* highest death rate: China; lowest death rate: Singapore
        
**Conclusion:**
* Singapore did quite well on testing (no. of testing) so that Singapore has high confirmed cases
* Taiwan did well on prevention (border control)
* China did not have good / sufficient medical equipment to handle the confirmed cases so that China has the highest death rate

```python
df_overview = pd.concat([hk.add_prefix('HongKong_'),
                         china.add_prefix('China_'),
                         taiwan.add_prefix('Taiwan_'),
                         korea.add_prefix('Korea_'),
                         japan.add_prefix('Japan_'),
                         malaysia.add_prefix('Malaysia_'),
                         singapore.add_prefix('Signapore_')],
                        axis=1)

ConfirmedCases_col = [col for col in df_overview.columns if 'ConfirmedCases' in col]
Fatalities_col = [col for col in df_overview.columns if 'Fatalities' in col]

df_overview_ConfirmedCases = df_overview[ConfirmedCases_col]
df_overview_Fatalities = df_overview[Fatalities_col]
```

```python
df_overview_ConfirmedCases.plot(figsize=(16,8),
                                title='ConfirmedCases Overview',
                                fontsize=12)
plt.show()
```

**Observation:**
* COVID19 outbreak in China first (Jan), followed by Korea (Mar)
* COVID19 outbreak in Singapore (Apr) probably because Singapore did not alert the pandemic in other countries

```python
df_overview_Fatalities.plot(#ax=plt.gca(),
                            figsize=(16,8),
                            title='Fatalities Overview',
                            fontsize=12)
plt.show()
```

**Observation:**
* The sudden increase in fatalities in China in Apr indicated that China suddenly changed the number of fatalities (incorporating the suspicious infected death cases)
